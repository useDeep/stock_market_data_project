```python
import json
from json import dumps, loads
from kafka import KafkaConsumer
from time import sleep
from s3fs import S3FileSystem
```

```python
consumer= KafkaConsumer(
    'stock',
    bootstrap_servers= ['100.26.137.53:9092'],
    api_version=(0,11,5),
    value_deserializer= lambda x: loads(x.decode('utf-8'))
)
```

```python
for c in consumer:
    print(c.value)
```

```python
s3= S3FileSystem(key='AWS_ACCESS_KEY', secret='AWS_SECRET')
```

```python
for count, i in enumerate(consumer):
    with s3.open("s3://20131a0332/stock_market_{}.json".format(count), 'w') as file:
        json.dump(i.value, file)
```
